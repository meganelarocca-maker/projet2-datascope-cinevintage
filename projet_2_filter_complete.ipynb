```python
import numpy as np
import pandas as pd

df_7["budget"] = pd.to_numeric(df_7["budget"], errors="coerce")


conditions = [
    df_7["budget"].isna(),
    df_7["budget"]==0,                  
    (df_7["budget"] < 500_000) & (df_7["budget"]>1),             
    (df_7["budget"] >= 500_000) & (df_7["budget"] < 3_000_000),  
    df_7["budget"] >= 3_000_000             
]


choices = ["unknown", "0", "indé pur", "cinéma indépendant", "blockbuster"]


df_7["budget_category"] = np.select(conditions, choices, default="unknown")

vc_budget=df_7["budget_category"].value_counts()

```

```python
lang_europe = {
    "de","es","pt","pl","sv","sk","it","ro","nl","da","no","is",
    "ru","cs","sl","hr","sr","be","uk","mk","bs","el","hy","sq", "et", "ca", "fi", "hu", "lv", "lt", "bg"
}

def classify(lang):
    if pd.isna(lang):
        return "unknown"
    if lang == "fr":
        return "français"
    if lang == "en":
        return "langue anglaise"
    if lang in lang_europe:
        return "européen"
    return "unknown"

df_7["origine"] = df_7["original_language"].apply(classify)

df_7["origine"].value_counts()

df_7["runtime"]
```
